```python
import pandas as pd
import twint
import nest_asyncio
```

```python
nest_asyncio.apply()
```

```python
conf=twint.Config()
```

```python
conf.Lang="fa"
conf.Limit=10000
conf.Since='2022-01-01'
conf.Until='2022-01-31'
conf.Output="2022-01.csv"
conf.Store_csv=True
```

```python
keys_txt=pd.read_csv('new_keys.txt')
```

```python
keys=keys_txt['keys'].str.strip()
```

```python
keys=keys.values.tolist()
```

```python
keywords=[]
for key in keys:
  keywords.append(key)
```

```python
for key in keywords:
  conf.Search=str(key)
  twint.run.Search(conf)
```
